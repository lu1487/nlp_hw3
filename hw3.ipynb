```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import re
import pandas as pd
import jieba
import jieba.analyse
```

```python
dm = pd.read_csv('movies_1.csv')
typecut = []
for d in dm['類型']:
    ch =re.compile("[\u4e00-\u9fa5]")
    seg_word =  "".join(ch.findall(d))
    top = jieba.lcut(seg_word)
    typecut.append(top)
namecut = []
for d in dm['中文名稱']:
    ch =re.compile("[\u4e00-\u9fa5]")
    name =  "".join(ch.findall(d))
    namecut.append(name)
tf = []
for j in dm['劇情介紹']:
    tf_temp = jieba.analyse.extract_tags(j,topK=5)
    tf.append(tf_temp)


```

```python
tf
```

```python
data = { 'type':typecut ,'name':namecut,'article':tf}
dm1 = pd.DataFrame(data)

dm1['type'] = dm1['type'].apply(lambda x :  str(x)[1:-2].replace("'","").replace(",",""))
dm1['article'] = dm1['article'].apply(lambda x :  str(x)[1:-2].replace("'","").replace(",",""))
dm1['key'] = dm1['name'].astype(str)+' '+dm1['type'].astype(str)+' '+dm1['article'].astype(str)
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(dm1['key'])
tfidf = TfidfTransformer() 
tf=tfidf.fit_transform(X)
word = vectorizer.get_feature_names() 
```

```python
number=500/6657
X_train, X_test, y_train, y_test = train_test_split(X, dm1['type'].str[0:2], test_size=number)
clf=KNeighborsClassifier(n_neighbors=51)
clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)    # predict 
y_test=y_test.values
score=metrics.accuracy_score(y_test, y_pred)

print("精準度：",score*100 ,"%")
```
